```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("../data/kopa_cesu_pils_apmekletaju_veidi_pa_gadiem_menesiem.csv", sep=";")
df
 
#1. Line Plot
 
dfg=df[df['Gads']==2014]
 
dfg.plot(x="Mēnesis", y="Skolēnu grupa", color="red")
plt.show()
 
#2. Multi-Line Plot
 
dfg=df[df['Gads']==2015]
dfg.plot(x="Mēnesis", y=["Skolēnu grupa", "Ģimene", "Draugi"] )
plt.show()
 
#3. Stabiņu Diagrammu
 
dfg=df[df['Gads']==2016]
dfg.plot.bar(x="Mēnesis", y=["Skolēnu grupa", "Ģimene", "Draugi"] )
plt.show()
 
#4. Pie-chart
 
labels = 'Janvāris', 'Februāris', 'Marts', 'Aprīlis', 'Maijs'
sizes = [576, 90, 257, 517, 972]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'black']
 
plt.title("Apmekletāju skaits 2014 gadā")
 
# Plot
plt.pie(sizes, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)
 
plt.axis('equal')
plt.show()
 
#5. Stack-Plot - aizpildītu laukuma diagrammu
dfg=df[df['Gads']==2016]
x = dfg["Mēnesis"]
 
y1 = dfg["Ģimene"]
y2 = dfg["Draugi"]
 
fig, ax = plt.subplots()
 
ax.stackplot(x, [y1, y2])
 
fig.set_figwidth(12)    
fig.set_figheight(6)    
fig.set_facecolor('floralwhite')
ax.set_facecolor('seashell')
 
plt.show()
```
